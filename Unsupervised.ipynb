```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs  # For generating sample data
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=4, random_state=42)
```

```python


df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
     
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
inertia = [] #list wcss value
#elbow method
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    
```

```python
plt.plot(range(1, 11), inertia, marker='o')#elbow graph
```

```python
kmeans = KMeans(n_clusters=10, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)
df['cluster'] = cluster_labels

sns.scatterplot(x=df['Feature_1'],
                y=df['Feature_2'],
                hue=df['cluster'],
                palette='viridis')
plt.title('K-Means Clustering Results')
```
