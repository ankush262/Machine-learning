```python
#Median
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
n = len(speed)
speed.sort()
mid =n//2
if n % 2 ==1:
    print(speed[mid])
else:
    print((speed[mid -1] + speed[mid])/2)
    

import numpy as np
x =np.median(speed)
print(x)

```

```python
#Mode
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
def mode(data):
    frequency = {}  # Dictionary to count occurrences
    
    # Count occurrences of each value
    for value in data:
        frequency[value] = frequency.get(value, 0) + 1
    
    # Find the maximum frequency
    max_freq = max(frequency.values())
    
    # Find all numbers that appear with max frequency
    modes = [key for key, val in frequency.items() if val == max_freq]
    
    return modes  # Return mode(s)

# Example usage
speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
print("Mode:", mode(speed))


from scipy import stats
x = stats.mode(speed)
print(x)
```

Variance
Variance is another number that indicates how spread out the values are.

In fact, if you take the square root of the variance, you get the standard deviation!

Or the other way around, if you multiply the standard deviation by itself, you get the variance!

```python
#variance
speed = [32,111,138,28,59,77,97]

def mean(speed):
    n = len(speed)

    total = 0
    for value in speed:
      total +=value
      mean = total/n
    return mean
mean_value = mean(speed)
squared_difference = []
for value in speed:
    difference = value - mean_value
    squared_difference.append(difference**2)

variance = mean(squared_difference)
print(variance)
    
#standard deviation
standard_deviation =variance**0.5
print(standard_deviation)
    
    
    
```
