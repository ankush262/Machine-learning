```python
#__str__() function
'''__str__() is called when print() or str(object) is used.
It should return a human-readable string representation.'''
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"{self.name}({self.age})"
    def Hello(self):
        print(f"HEllo my name is {self.name} and my age is {self.age} ")
p1 =  Person("john",20)       

print(p1,p1.Hello())
#modfiy object properties
p1.age = 50
#delete objects
del p1
```

Iterators  - Iterators is just like for loop with __iter()__ for current iteam and next() method for upcoming value to repeat

The __iter__() method acts similar, you can do operations (initializing etc.), but must always return the iterator object itself.

The __next__() method also allows you to do operations, and must return the next item in the sequence.

```python
mytuple = ("apple", "banana", "cherry")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))

#create an iterator
class MYnumbers:
    def __iter__(self):
        self.a = 1
        return self
    def __next__(self):
        x = self.a
        self.a += 1
        return x

myclass = MYnumbers()
myiter = iter(myclass)
print(next(myiter))
print(next(myiter))


#stop Iteration
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
  print(x)
```
