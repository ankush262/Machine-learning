LISTS in python

```python
# Example of a list
my_list = ["apple", "banana", "cherry"]

# Accessing the first item
print(my_list[0])  # Output: apple

# Changing the second item
my_list[1] = "blueberry"
print(my_list)  # Output: ['apple', 'blueberry', 'cherry']

# Looping through the list
for item in my_list:
    print(item)
#compression in list
above = [item for item in my_list]
# Getting the length of the list
print(len(my_list))  # Output: 3


# Adding an item to the list
my_list.append("date")
print(my_list)  # Output: ['apple', 'blueberry', 'cherry', 'date']

# Removing an item from the list
my_list.remove("blueberry")
print(my_list)  # Output: ['apple', 'cherry', 'date']

#List Methods Python lists have several built-in methods, such as append(), remove(), pop(), sort(), and more.

# Using the sort() method
my_list.sort()
print(my_list)  # Output: ['apple', 'cherry', 'date']
my_list.pop(2)
print(my_list)  #output :['apple', 'cherry']

# Creating a list using list comprehension
squares = [x**2 for x in range(10)]
print(squares)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

SETS in python

```python
#create a set
thisset = {"apple","cherry","banana"}
print(type(thisset))
print(len(thisset))
#set constructor
thisset =set(("apple","banana","cherry"))
'''You cannot access items in a set by referring to an index or a key.
But you can loop through the set items using a for loop, or ask if a specified value is present in a set, by using the in keyword'''
for x in thisset:
  print(x)

print("banana" not in thisset)

#add set items
thisset.add("orange") #output : banana apple cherry orange
tropical = {"pineapple", "mango", "papaya"}
thisset.update(tropical)
print(thisset)
#remove(), discard(),pop(),clear(),del()
thisset.remove("mango")
thisset.discard("strawberry")
thisset.pop()
print(thisset)
thisset.clear()
print(thisset)
del thisset
```
